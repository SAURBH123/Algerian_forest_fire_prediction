```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Algerian_clean_dataset.csv')
```

```python
df.head()
```

```python
df["Classes"]=df['Classes'].map({'not fire':0,'fire':1})
```

```python
df.head()
```

```python
df.drop(['Unnamed: 0','day','month','year'],axis=1,inplace=True)
```

```python
df.head()
```

```python
df.to_csv('Algerian_forest_fire_dataset.csv',index=False)
```

```python
y=df['FWI']
```

```python
X=df.drop('FWI',axis=1)
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
X_train.shape,X_test.shape
```

```python
df.corr()
```

```python
X_train.corr()
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(X_train.corr(),annot=True)
```

```python
def correlation(dataset, threshold):
    col_corr = set()
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i, j]) > threshold: 
                colname = corr_matrix.columns[i]
                col_corr.add(colname)
    return col_corr
```

```python
correlation(X_train,0.85)
```

```python
X_train.drop(['BUI','DC'],axis=1,inplace=True)
X_test.drop(['BUI','DC'],axis=1,inplace=True)
```

```python
X_train.shape,X_test.shape
```

```python
df
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
X_train_scaled=scaler.fit_transform(X_train)
```

```python
X_test_scaled=scaler.transform(X_test)
```

```python
axis,fig=plt.subplots(1,2,figsize=(15,8))
plt.subplot(121)
sns.boxplot(X_train)
plt.subplot(122)
sns.boxplot(X_train_scaled)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error,mean_squared_error
from sklearn.metrics import r2_score
regresor=LinearRegression()
regresor.fit(X_train_scaled,y_train)
y_predict=regresor.predict(X_test_scaled)
mse=mean_squared_error(y_test,y_predict)
mar=mean_absolute_error(y_test,y_predict)
score=r2_score(y_test,y_predict)
adjusted_r2=1-((1-score)*(len(y_test)-1)/(len(y_test)-len(X_test.columns)-1))
print(regresor.coef_)
regresor.intercept_
```

```python
mse,mar,score,adjusted_r2
```

```python
from sklearn.linear_model import Lasso
from sklearn.metrics import mean_absolute_error,mean_squared_error
from sklearn.metrics import r2_score
lasso=Lasso()
lasso.fit(X_train_scaled,y_train)
y_predict=lasso.predict(X_test_scaled)
mse=mean_squared_error(y_test,y_predict)
mar=mean_absolute_error(y_test,y_predict)
score=r2_score(y_test,y_predict)
adjusted_r2=1-((1-score)*(len(y_test)-1)/(len(y_test)-len(X_test.columns)-1))
print(lasso.coef_)
lasso.intercept_
```

```python
mse,mar,score,adjusted_r2
```

```python
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_absolute_error,mean_squared_error
from sklearn.metrics import r2_score
ridge=Ridge()
ridge.fit(X_train_scaled,y_train)
y_predict=ridge.predict(X_test_scaled)
mse=mean_squared_error(y_test,y_predict)
mar=mean_absolute_error(y_test,y_predict)
score=r2_score(y_test,y_predict)
adjusted_r2=1-((1-score)*(len(y_test)-1)/(len(y_test)-len(X_test.columns)-1))
print(ridge.coef_)
ridge.intercept_
```

```python
mse,mar,score,adjusted_r2
```

```python
from sklearn.linear_model import ElasticNet
from sklearn.metrics import mean_absolute_error,mean_squared_error
from sklearn.metrics import r2_score
elastic_net=ElasticNet()
elastic_net.fit(X_train_scaled,y_train)
y_predict=elastic_net.predict(X_test_scaled)
mse=mean_squared_error(y_test,y_predict)
mar=mean_absolute_error(y_test,y_predict)
score=r2_score(y_test,y_predict)
adjusted_r2=1-((1-score)*(len(y_test)-1)/(len(y_test)-len(X_test.columns)-1))
print(elastic_net.coef_)
elastic_net.intercept_
```

```python
mse,mar,score,adjusted_r2
```
